```python
import magpy
import matplotlib.pyplot as plt
import skimage.io as sk_i
from importlib import reload  
import numpy as np
from scipy.optimize import curve_fit
from scipy.special import voigt_profile
from scipy.ndimage import convolve1d
import plasmapy
from Thomson import S_k_omega
%matplotlib notebook
```

```python
sh = 's0903_20 TS shot.asc'
bk = 'bk.asc'
```

```python
reload(magpy)
spect_sh = magpy.Spectrometer(sh)
spect_bk = magpy.Spectrometer(bk)
bundle = magpy.Bundle()
bundle.sp=17.9
bundle.off=7.7
bundle.dis=2
bundle.l0=531
bundle.l1=533
bundle.dl = 529.
bundle.split(spect_sh, spect_bk)
f7 = bundle.fibers['7A']
```

```python
fig, ax = plt.subplots()
spect_bk.imshow(ax, cmap='Reds')
```

```python
bundle.draw_split(spect_bk)
```

```python
fig, ax = plt.subplots()
ax.scatter(f7.b_l, f7.b_y)
```

```python
f7.fit_response()
f7.setup_convolution_grid(0.1)
x_fit = np.linspace(531, 533, 500)
y_fit = f7.get_response(x_fit)
fig, ax = plt.subplots()
ax.plot(x_fit, y_fit)
ax.scatter(f7.b_l, f7.b_y)
```

```python
l0 = f7.get_probe_wavelength()*1e-9
S, alpha = S_k_omega(f7.l*1e-9, l0, 90., 14, 20, 15, 1e18, 6,  v_fi=0, v_fe=0)
kernal = f7.kernal
kernal /= kernal.max()
S_conv = convolve1d(S, kernal, mode='constant')
S_conv /= S_conv.max()
```

```python
fig, ax = plt.subplots()
ax.plot(f7.l, S, label='unconvolved', lw=1)
#ax.plot(f7.kernal, label='response', marker='o')
ax.plot(f7.l, S_conv, label='convolved',lw=1)

ax.legend()
```
